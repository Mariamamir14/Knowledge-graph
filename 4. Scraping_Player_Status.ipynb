```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
Players = len(df.full_name.unique()) #5827
RealgmURL = len(df.ID.unique())#5886
Players
```

```python
df = df.dropna()
```

```python
df = pd.DataFrame({
    'full_name':df.full_name.unique(),
    'ID':df.ID.unique(),
    'player':df.player.unique(),
})
df
```

```python
# URL of the RealGM player summary page for LeBron James
#Here I give all the players url to get their contract
Contracts = []
for url in df.ID:
    response = requests.get(url)
    content = response.content
    soup = BeautifulSoup(content, 'html.parser')
    nba_status_element = soup.find('div', {'class': 'half-column-right'})
    Contracts.append(nba_status_element.find('p').text.replace('Current NBA Status: ',''))
df['Contracts'] = Contracts
```
